## 1.2 Генеральная совокупность и выборка

### Основные определения:

* Генеральная совокупность (далее ГС) — совокупность всех объектов, относительно которых предполагается делать выводы при изучении конкретной задачи;  

* Выборка - часть ГС;
* Репрезентативная выборка – выборка, которая отражает основные свойства ГС.

### Способы формирования репрезентативной выборки

1) Простая случайная выборка (simple random sample) - случайным образом выбираются элементы ГС;
2) Стратифицированная выборка (stratified sample) – разделение ГС на обособленные страты, затем из этих страт случайным образом выбираются элементы;  
* Выборка формируется из всех страт;
* В пределах страты элементы должны быть однородны, а между стратами должны быть различия.
3) Групповая выбока (cluster sample) - ГС делится на несколько кластеров, из нескольких кластеров случайным образом выбираются элементы;
* В пределах кластера элементы должны быть разнородны, тогда как поддерживается схожесть между разными кластерами;
* Повышает эффективность выборки, уменьшая стоимость (сложность).

### Примеры:


  

**А** - Простая случайная выборка (random sample)  
**Б** - Статифицированная выборка (stratified sample)  
**В** - Групповая выборка (cluster sample)

## 1.4 Меры центральной тенденции

### Основные определения:

Типы описательной статистики:
1) Меры центральной тенденции - позволяет определить, насколько *высокие* значения принимает распределение;
2) Меры изменчивости - вариативность распределения.

**Мода** - это значение признака, которое встречается максимально часто. В выборке может быть несколько мод.  

**Медиана** - это значение признака, которое делит упорядочное множество пополам. Если множество содержит чётное количество элементов, то берётся среднее из двух серединных элементов упорядочного множества.

**Среднее значение** - сумма всех значений измеренного признака делённое на количество измеренных значений.

### Свойства среднего значения
1) Если к каждому значению
выборки прибавить число **c**, то среднее значение увеличится на это число;
2) Если каждое значению выборки умножить на число **c**, то среднее значение увеличится в **с** раз;
3) Если для каждого значения выборки рассчитать отклонение от среднего арифметического, то сумма таких отклонений будет = 0.

### Примеры:

*Красная линия* - Мода  
*Зеленая линия* - Медиана  
*Синяя линия* - Среднее значение

```python
'''Самостоятельный расчет моды, медианы и среднего'''
nums = [int(i) for i in '185 175 170 169 171 175 157 172 170 172 167 173 168\
        167 166 167 169 172 177 178 165 161 179 159 164 178 172 170 173 171'.split()]

def get_mode(num_list):
    counts_of_num = {}
    for i in num_list:
        if i in counts_of_num:
            counts_of_num[i] += 1
        else:
            counts_of_num[i] = 1
            
    count = max(counts_of_num.values())         
    mode = max(counts_of_num, key=lambda key: counts_of_num[key])
    
    return mode, count
           

def get_median(num_list):
    median = 0
    
    num_list = sorted(num_list)
    
    if (len(num_list) / 2) % 1 == 0:
        median += (num_list[len(num_list) // 2 - 1] + num_list[len(num_list) // 2]) / 2
    else:
        median += (num_list[len(num_list) // 2])
        
    return median


def get_mean(num_list):
    mean = sum(num_list) / len(num_list)
    
    return mean


print('Mode =', get_mode(nums)) 
print('Median =', get_median(nums)) 
print('Mean =', get_mean(nums))
```

## 1.5 Меры изменчивости

### Основные определения:

**Размах** - разность между максимальным и минимальным значениям выборки. Крайне чувствителен к выбросам.  

**Дисперсия** - средний квадрат отклонений индивидуальных значений признака от их средней величины. При расчете дисперсии для выборки от знаменателя формулы вычитается 1.


**Среднеквадратическое отклонение** - корень из дисперсии (реальное среднее отклонение от среднего по выборке)

### Свойства  стандартного отклонения и дисперсии
1) Если к каждому наблюдению выборки прибавить число **c**, то ни дисперсия, ни стандартное отклонение не изменятся;  
2) Если каждое число выборки умножить на число **с**, то стандартное отклонение изменится в **c** раз, а дисперсия в **c^2** раз.

```python
'''Самостоятельный расчет размаха, дисперсии и стандартного отклонения'''
def get_range(num_list):
    min_num = min(num_list)
    max_num = max(num_list)

    nums_range = max_num - min_num
    return nums_range

def get_dispersion(num_list, ddof=0):         # где ddof - степень свободы
    mean = sum(num_list) / len(num_list)
    m_deviation = 0
    
    for i in num_list:
        m_deviation += (i - mean)**2
    
    D = m_deviation / (len(num_list) - ddof)  
    return D

def get_standart_deviation(num_list, ddof=0): # где ddof - степень свободы
    mean = sum(num_list) / len(num_list)
    m_deviation = 0
    
    for i in num_list:
        m_deviation += (i - mean)**2
    
    D = m_deviation / (len(num_list) - ddof)
    std = D**0.5   
    return std

nums = [int(i) for i in '1 5 2 7 1 9 3 8 5 9'.split()]

print('Range =', get_range(nums))
print('D =', get_dispersion(nums, 1))
print('std =', get_standart_deviation(nums, 1))
```

```python
'''Расчет дисперсии и стандартного отклонения с помощью numpy'''
import numpy as np

nums = [int(i) for i in '1 5 2 7 1 9 3 8 5 9'.split()]

print('D =', np.var(nums, ddof=1))
print('std =', np.std(nums, ddof=1)) #Учитываем степень свободы - ddof
```

```python
'''Расчет дисперсии и стандартного отклонения с помощью pandas'''
import pandas as pd

sample_1 = [1, 3, 5, 6, 6, 7, 9, 11]
sample_2 = [5, 7, 9, 10, 10, 11, 13, 15]

df = pd.DataFrame({
    "Выборка 1": sample_1,
    "Выборка 2": sample_2
})

print("Дисперсия\n", df.var())
print("Cтандартное отклонение \n", df.std(ddof=1))
```

## 1.7 Нормальное распределение

### Основные определения:  

**Нормальное распределение** - унимодальное, симметричное распределение. Отклонение наблюдений от среднего подчиняются определенному вероятностному закону.

**Правило двух и трех сигм** - в соответствии с вероятностным законом:
* 68% наблюдений находятся в диапазоне +-σ;
* 95% наблюдений находятся в диапазоне +-2σ;  
* 100% наблюдений находятся в диапазоне +-3σ.  

Нормальное распределение возникает в результате воздействия множества факторов, вклад каждого из которых очень мал. Для облегчения этого восприятия в 1873 году Фрэнсис Гальтон сделал устройство, которое в последствии назвали Доской Гальтона (или квинкункс). Суть простая: сверху по середине подаются шарики, которые при прохождении нескольких уровней на каждом уровне сталкиваются с препятствием, и при каждом столкновении отскакивают либо влево, либо вправо (с равной вероятностью). 

```python
'''Иммитация доски Гальтона в коде'''
import seaborn as sns
data = dict()
# количество шариков
N = 10000
# количество уровней
level = 20
for _ in range(N):
    index = 0
    for _ in range(level):
        index += np.random.choice([-1, 1])
    data.setdefault(index, 0)
    data[index] += 1
sns.barplot(x=list(data.keys()), y=list(data.values()));
```

**Z-преобразование (стандартизация)** - преобразование полученных данных в стандартную Z-шкалу (Z-scores) со средним значением Mz = 0 и дисперсией Dz = 1. Чтобы привести к такому виду из каждого наблюдения нужно отнять среднее значение и разделить на стандартное отклонение:

При Z-преобразовании форма распределения не изменяется.  
Если отдельное наблюдение меньше, чем выборочное среднее, то соответствующее z - значение будет *отрицательным*.  
Если для некоторого наблюдения z значение равняется нулю, следовательно это наблюдение совпадает со средним значением по выборке.

### Примеры:  
Допустим, мы измерили рост 1000 человек, данное распределение оказалось нормальным со средним равным 175 и стандартным отклонением равным 8 (M = 175, sd = 8). Рост одного из испытуемых составил 186.2. Чему равняется z - значение, рассчитанное для этого испытуемого?

```python
x = 186.2    # рост испытуемого
M = 175      # среднее
sd = 8       # стандартное отклонение

z = (x-M)/sd # z-значение испытуемого
z
```

С помощью **Z-преобразования** можно ответить на вопрос: какой процент наблюдений лежит в определенном диапозоне?

```python
import scipy.stats as st

x_min = 154  
x_max = 158     # целевые значения
M = 150         # среднее значение выборки
sd = 8          # стандартное отклонение

z = (x_min - M) / sd # z-преобразование
print("Процент наблюдений выше 154 =", (1 - st.norm.cdf(z)) * 100)


'''Еще вычисления с помощью scipy'''

print("Случайное значение из распределения -", st.norm.rvs(M, sd))
# Survival Function 
print("Процент наблюдений выше 154 -", st.norm.sf(x_min, M, sd) * 100)  
# Cumulative Density Function
print("Процент наблюдений ниже 158 -", st.norm.cdf(x_max, M, sd) * 100)  
# Percent Point Function
print("Значение, до которого находятся 99% наблюдений -", st.norm.ppf(0.99, M, sd)) 
```

```python
'''Допустим, что некоторый признак распределен нормально, выборочное среднее
равняется 100, а дисперсия равняется 25 (M = 100, D = 25).
Тогда приблизительно 95% всех наблюдений находится в диапазоне: 
'''

M = 100
D = 25
sd = D**0.5
# Поскольку 95% от всех значений находятся в диапазоне +-2σ, то мин/макс занчения z = +-2
z_min = -2
z_max = 2

# z = (x-M)/sd
x_min = z_min*sd + M
x_max = z_max*sd + M
print('Приблизительно 95% всех наблюдений находится в диапазоне ', [x_min, x_max])
```

```python
''' Считается, что значение IQ (уровень интеллекта) у людей имеет нормальное распределение
со средним значением равным 100 и стандартным отклонением равным 15 (M = 100, sd = 15).
Какой приблизительно процент людей обладает IQ > 125?
'''

M = 100
sd = 15
x_min = 125

z = (x_min - M) / sd 
print("Процент людей с IQ выше 125 =", round((1 - st.norm.cdf(z)) * 100))
```

```python
'''Считается, что значение IQ (уровень интеллекта) у людей имеет нормальное распределение
со средним значением равным 100 и стандартным отклонением равным 15 (M = 100, sd = 15).
Какой приблизительно процент людей обладает IQ  на промежутке от 70 до 112 '''

M = 100
sd = 15

x_min = 70
x_max = 112

z_min = (x_min - M) / sd
z_max = (x_max - M) / sd

print("Процент людей с IQ выше 70 и ниже 112 =", round((1 - st.norm.cdf(z_min)) * 100 - (1 - st.norm.cdf(z_max)) * 100))
```

## 1.8 Центральная предельная теорема (ЦПТ)

### Основные определения:

**ЦПТ** - множество средних выборок из генеральной совокупности (ГС необязательно иметь нормальное распределение) будут иметь нормальное распределение. Причём, среднее значение этих выборок будет близко к среднему генеральной совокупности, а **стандарное отклонение (sd)** выборок будет называться **стандарной ошибкой среднего (se)** и отображает насколько в среднем выборочное значение откланяется от среднего ГС.  

**Стандартная ошибка среднего (se)** - это среднеквадратическое отклонение распределения выборочных средних.  
Чем больше число наблюдений и чем меньше изменчивость признаков ГС, тем меньше **se**. Чем меньше  **se**, тем реже выборочные средние будут сильно отклоняться от среднего в ГС.

Зная **стандартное отклонение ГС (σ)** и **размер выборки (N)** можно рассчитать **стандартную ошибку среднего (se)**:  
  

Если размер выборки достаточно большой (>30) и она является репрезентативной, то вместо σ берется **sd**:  


### Проверка ЦПТ:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

# Формирование ГС
# значения игральной кости
dice = [1, 2, 3, 4, 5, 6]
# количество бросков кости
count = 6
# размер ГС
# sp - Statistical population - генеральная совокупность
sp_size = 10000
sp = pd.Series(dtype=np.int64, index=range(sp_size))
for i in range(sp_size):
    value = 0
    for _ in range(count):
        value += np.random.choice(dice)
    sp[i] = value

sp.plot.hist(bins=28)
```

```python
'''Распределение выборочных средних является нормальным, со средним =(стремящимся к) среднему зачению в ГС'''
# формирование выборок из ГС
# количество выборок
samples_count = 100
# размер выборки
sample_size = 100
samples = pd.DataFrame([[np.random.choice(sp) for _ in range(sample_size)] for __ in range(samples_count)]).T

samples.hist(figsize=(16, 10), sharex=0)
plt.subplots_adjust(hspace = 0.6)
```

```python
means = samples.mean()
print('Cреднее ГС:', sp.mean(), '; Cреднее средних выборок:',  means.mean())
print('Разница:', abs(means.mean() - sp.mean()), '; Стандартная ошибка среднего:', means.std())
```

```python
# произвольная выборка 
sample = samples[42]
print('sample mean:', sample.mean())
print('sample std:', sample.std())
print('sample se: ', sample.std()/math.sqrt(sample.size))
```

```python
samples = samples.T
N = math.sqrt(sample_size)
max_ = sample_size - 1

samples['mean'] = [samples.loc[i, :max_].mean() for i in range(samples_count)]
samples['std'] = [samples.loc[i, :max_].std() for i in range(samples_count)]
samples['se'] = [(samples.loc[i, :max_].std()/N) for i in range(samples_count)]
samples
```

```python
sp.plot.hist(bins=28)
```

```python
samples['mean'].plot.hist(bins=10)
```

```python
print("Среднее средних выборок",samples['mean'].mean())
print("Среднее ГС", sp.mean())
```

## 1.9 Доверительные интервалы для среднего

### Основные определения:
**Доверительные интервалы** - интервал, в пределах которого с заданной вероятностью лежат выборочные оценки статистических характеристик генеральной совокупности. Другими словами, если имеется некоторая выборка и ГС, то нельзя точно знать среднее значение ГС, зная только среднее выборки. Однако, с некоторым процентом уверенности, можно сказать в каком интервале лежит среднее ГС.  
В центре доверительного интервала находится выборочное среднее значение **(x_mean)**.
При увеличении объема выборки, доверительный интервал становится более *узким*.


X4 - доверительный интервал, который не включает в себя среднее ГС, поскольку его выборочное среднее поскольку находится далеко от среднего ГС.

### Примеры:

```python
"""В исследовании приняла участие выборка из 64 людей(N) в возрасте [18:30] лет. 
Имеются результаты исследования (экспрессия гена) со средним(M) = 100 и стандартный отклонением(sd) = 4.
Чему равно среднее значение экспрессии гена в ГС(все люди в возрасте от 18 до 30 лет)?"""

x_mean = 100
sd = 4
N = 64

se = sd / N**0.5

# рассчет 95% доверительного интервала
# x_mean-1.96se|-------x_mean-------|x_mean+1.96se
# 1,96 это количество сигм  в нормальном распределении, необходимые, чтобы охватить 95% значений в этом распределении

""" Если многократно повторять эксперимент, для каждой выборки рассчитывать свой доверительный интервал,
    то в 95 % случаев истинное среднее будет находиться внутри доверительного интервала."""

print("95% Доверительный интервал ", [x_mean - 1.96*se, x_mean + 1.96*se])
```

```python
"""Рассчитайте 99% доверительный интервал для следующего примера: """

x_mean = 10
sd = 5
N = 100

se = sd / N**0.5
print("99% Доверительный интервал ", [x_mean - 2.58*se, x_mean + 2.58*se])
```

```python
"""Вычисление sigma c помощью scipy"""
from scipy import stats

# 95%
p = 0.95
# так как интервал двухсторонний , выполняется вычисление
alpha = (1-p)/2
# isf - Inverse survival function (inverse of sf) 
print(f'{stats.norm().isf(alpha):.2f} sigma')

# 99%
p = 0.99
alpha = (1-p)/2
print(f'{stats.norm().isf(alpha):.2f} sigma')
```

```python
""" Решение задачи с помощью scipy
Рассчитайте 99% доверительный интервал для следующего примера: 
среднее = 10, стандартное отклонение = 5, размер выборки = 100"""

from scipy import stats

p = 0.99
x_mean = 10
sd = 5
N = 100

# стандарная ошибка среднего
se = sd / N**0.5

# sigma
alpha = (1-p)/2
sigma = stats.norm().isf(alpha)

# доверительный интервал
сonfidence_interval = x_mean - sigma*se, x_mean + sigma*se

print('[%.2f, %.2f]' % сonfidence_interval)
```

## 1.10 Идея статистического вывода, p-уровень значимости

### Основные идеи статистического вывода:  
Сначала допускается, что верна **нулевая (H0) гипотеза** (нет никаких различий с ГС).  
После рассчитывается вероятность того, что различия получены абсолютно случайно . 
* При помощи **p** (п-уровень значимости) выявляется, какую гипотезу будет считаться наиболее правильной;
* Чем меньше **p**, тем больше оснований отклонить H0 (обычно берут **p** < 0.05);
* **p** - вероятность события, если допустить, что H0 гипотеза верна;
* Сам **p** ничего не говорит ни о правильности, ни о ценности, ни о научности получаемых результатов.

**Ошибки статистического вывода:**  
* Ошибка первого рода - отклонена H0, когда она была верна;
* Ошибка второго рода - не отклонена H0, когда верна H1.

### Примеры:

```python
"""Известно, что на выздоровлении при некотором заболевании в среднем требуется 20 дней (M=20).
   Был разработан новый препарат и нужно проверить можно ли сократить этот срок.
   Была набрана выборка из 64 людей (N=64) и на них был опробован новый метод лечения.
   Средний срок выздоровления сократился до 18.5(x_mean=18.5). Стандартное отклонение = 4."""
M = 20
N = 64
x_mean = 18.5
sd = 4


# Нулевая гипотеза(H0) - никакого воздействия препарат не оказывает и среднее ГС тех? кто использует препарат = 20 (M_ud = 20).
# Альтернативная гипотеза(H1) - препарат влияет. M_ud ≠ 20.

# Предполагается, что верна H0
# в соответствии с ЦПТ, x_mean будут распределены вокруг M_ud
se = sd/ N**0.5         # se = 0.5
# вычисляется насколько далеко x_mean отклонилось от M_ud(предпологаемое среднее ГС) в единицах стандартного отклонения
z = (x_mean - M) / se   # z = -3
# это означает, что если бы M_ud = 20, то x_mean - отклонилось бы на -3σ в левую сторону.
# рассчитывается вероятность такого или еще более сильного отклонения x_mean от M_ud.
# рассчет через онлайн-калькулятор Distribution Calculator https://gallery.shinyapps.io/dist_calc/
# вероятность отклонения в левую и правую сторону (двусторонний p-уровень значимости) p = 0.0027
# Чем меньше p, тем больше оснований отклонить H0 (обычно берут p < 0.05)
# В таком случае нулевая гипотеза отклоняется. 
# если p > 0.05 , то недостаточно оснований отклонить H0.
```

```python
"""Использование доверительных интервалов зачастую рассматривают, как альтернативный способ проверки гипотез.
   В нашем случае, если значение 20 (предполагаемое среднее значение в генеральной совокупности)
   не будет принадлежать 95% доверительному интервалу, рассчитанному по выборочным данным,
   у нас будет достаточно оснований отклонить нулевую гипотезу. Проверьте, согласуются ли результаты
   двух этих подходов: рассчитайте 95% доверительный интервал для среднего значения,
   на примере с тестированием нового препарата. N = 64,  sd = 4,   x_mean = 18.5"""

from scipy import stats

N = 64
sd = 4
x_mean = 18.5
p = 0.95

M = 20


se = sd/ N**0.5

alpha = (1-p)/2
sigma = stats.norm().isf(alpha)


сonfidence_interval = x_mean - sigma*se, x_mean + sigma*se

print("20 не принадлежит доверительному интервалу [%.2f, %.2f] — отклоняем Н0" % сonfidence_interval)
```

```python
"""Вычисление p-уровня с помощью scipy.
   В среднем слушатели курса по введению в статистику набирают 115 баллов, однако, в 2015 году средний балл
   случайно выбранных 144 участников составил 118 со стандартным отклонением равным 9. Рассчитайте p уровень
   значимости для проверки нулевой гипотезы о том, что среднее значение баллов в 2015 году равняется 115.
"""

from scipy import stats

N = 144 
sd = 9 
x_mean = 118 

M = 115 

se = sd /  N**0.5 

z = (x_mean - M) / se

p = 2*(1-stats.norm.cdf(z))
print("p уровень = ", p, " < 0.05 - отклоняем гипотезу")
```
