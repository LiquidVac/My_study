## 1.3 Типы переменных. Количественные и номинативные переменные

### Определения:

Типы переменных:
1) Количественные:  
* Непрерывные – переменная принимает любое значение на промежутке (напр.[160; 190] 175.5);  
* Дискретные – только определенные значения (1,2,3,4 - 3.5 человека не бывает).  

2) Качественные:
* Номинативные – используются как маркеры для групп(например: 1 - женщины, 2 – мужчины). 
* Ранговые – используются для сравнения (1-ое место быстрее второго, а третье медленнее и т.д.)
